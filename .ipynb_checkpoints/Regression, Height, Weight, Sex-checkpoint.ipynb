## What we are going to do

We will use a simple data set consisting of three variables: height, weight, and sex. We will
    use weight as the dependent variable and height and sex as the independent variables.
    We will fit four different  models and compare them.

## Import modules

```python
import pandas as pd
import numpy as np

import statsmodels.api as sm
import statsmodels.formula.api as smf

import matplotlib.pyplot as plt
import seaborn as sns
```

## Get data

```python
df01 = pd.read_csv("ht-wt-sex-data.csv")
print(df01.columns)
df01.head()
```

## Use dummy variable

It appears that *Sex* is non-numeric. We will convert it to a numeric variable by 
creating a dummy variable: Male == 1 and Female  == 0.

```python
df02 = pd.get_dummies(df01, columns = ['Sex'], drop_first = True)
df02.head()
```

## Describing the data

Notice that the metrmics for the dummy variable are meaningless.

```python
df02.describe()
```

## Examining the data graphically

We can use various types of graphics to examine the distribution of the data. We can see that the
distribution of the height and weight variables are somewhat normal, but that the scatter plot
is very noisy. We can't expect our predictors to explain much of the variance in this dataset.

```python
x = df01['Height']
y = df01['Weight']
#x.plot.density()
fig, ax = plt.subplots(1,3,figsize = (18,6))
sns.kdeplot(x, ax=ax[0])
ax[0].set_title("Height, KDE Density Plot")
ax[0].set_xlabel("Height, in inches")
sns.kdeplot(y, ax=ax[1])
ax[1].set_title("Weight, KDE Density Plot")
ax[1].set_xlabel("Weight, in pounds")
ax[2].scatter(x, y, s=60, alpha=0.7, edgecolors="k")
ax[2].set_title("Weight (Y) ~ Height (X)");
ax[2].set_xlabel("Height, in inches");
ax[2].set_ylabel("Weight, in pounds");
```

## Looking at the distrbution for sex

Males are taller and weigh more than females. This difference is more apparent with
heght than with weight.

```python
fig, ax = plt.subplots(1, 2,figsize = (12,6))
sns.boxplot(data=df02, y="Height", x="Sex_Male", ax = ax[0])
sns.boxplot(data=df02, y="Weight", x="Sex_Male", ax = ax[1])
```

## First model

We predict weight using height only.

```python
model01 = smf.ols("Weight ~ Height", df02)
result01 = model01.fit()
result01.summary()
```

## Second model

We predict weight using sex only.

```python
model02 = smf.ols("Weight ~ Sex_Male", df02)
result02 = model02.fit()
result02.summary()
```

## Third model

We predict weight using height and sex together.

```python
model03 = smf.ols("Weight ~ Height + Sex_Male", df02)
result03 = model03.fit()
result03.summary()
```

## Fourth model

We predict weight using height, sex, and the interaction between height and sex.

```python
model04 = smf.ols("Weight ~ Height + Sex_Male + Height*Sex_Male", df02)
result04 = model04.fit()
result04.summary()
```

## Compare R-Square

```python
print("Model 01 R Square:", result01.rsquared)
print("Model 02 R Square:", result02.rsquared)
print("Model 04 R Square:", result03.rsquared)
print("Model 01 R Square:", result04.rsquared)

```
